```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load image in grayscale
img = cv2.imread('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/C.png', cv2.IMREAD_GRAYSCALE)

threshold_values = [0, 50, 100, 150, 200]

fig, axes = plt.subplots(1, len(threshold_values) + 1, figsize=(18, 4))
axes[0].imshow(img, cmap='gray')
axes[0].set_title('Original')
axes[0].axis('off')

for i, t in enumerate(threshold_values):
    ret, thresh = cv2.threshold(img, t, 255, cv2.THRESH_BINARY)
    axes[i+1].imshow(thresh, cmap='gray')
    axes[i+1].set_title(f'T = {t}')
    axes[i+1].axis('off')

plt.tight_layout()
plt.savefig('threshold_comparison.png', dpi=150)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from skimage import data, img_as_float
from skimage import exposure
```

```python
#Step 1: Load the moon image
img = data.moon()
```

```python
#Step 2: Contrast Stretching
# Rescale intensity values to include all intensities within the 2nd and 98th percentiles
p2, p98 = np.percentile(img, (2, 98))
img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))
```

```python
#Step 3: Display the image with its histogram

fig = plt.figure(figsize=(10, 7))
fig.add_subplot(2, 1, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Original image')
fig.add_subplot(2, 1, 2)
plt.hist(img.flat, bins=256, range=(0, 255))
plt.xlabel('gray level values')
plt.ylabel('Number of pixels')
plt.show()


fig = plt.figure(figsize=(10, 7))
fig.add_subplot(2, 1, 1)
plt.imshow(img_rescale, cmap='gray')
plt.axis('off')
plt.title('Contrast stretch')
fig.add_subplot(2, 1, 2)
plt.hist(img_rescale.flat, bins=256, range=(0, 255))
plt.xlabel('gray level values')
plt.ylabel('Number of pixels')
plt.show()
```
