```python
import numpy as np
```

```python
# 1. Increase the size of the image (Scale Up)
img = cv2.imread('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/images.jpg')

# Scale factor: 2x width and height
scaled_img = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)

cv2.imshow('Original Image', img)
cv2.waitKey()

cv2.imshow('Scaled Up Image (2x)', scaled_img)
cv2.waitKey()

cv2.imwrite('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/scaled_image.jpg', scaled_img)
```

```python
# 2. Rotate the image by 120 degrees
img = cv2.imread('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/images.jpg')

# Get image center
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

# Build rotation matrix: 120 degrees, scale = 1.0
M_rotate = cv2.getRotationMatrix2D(center, 120, 1.0)

# Warp (rotate) the image
rotated_img = cv2.warpAffine(img, M_rotate, (w, h))

cv2.imshow('Rotated Image (120 degrees)', rotated_img)
cv2.waitKey()

cv2.imwrite('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/rotated_image.jpg', rotated_img)
```

```python
# 3. Shear operation
img = cv2.imread('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/images.jpg')

(h, w) = img.shape[:2]

# Shear factor along x-axis
shear_factor = 0.3

# Shear transformation matrix
M_shear = np.float32([
    [1, shear_factor, 0],
    [0, 1,            0]
])

# Calculate new width to avoid cropping
new_width = w + int(h * shear_factor)

# Apply shear using warpAffine
sheared_img = cv2.warpAffine(img, M_shear, (new_width, h))

cv2.imshow('Sheared Image', sheared_img)
cv2.waitKey()

cv2.imwrite('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/sheared_image.jpg', sheared_img)
```

```python
# 1. Negative Transformation - inverts pixel values using s = 255 - r
img = cv2.imread('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/images.jpg')
negative_img = 255 - img
cv2.imshow('Original Image', img)
cv2.waitKey()
cv2.imshow('Negative Image', negative_img)
cv2.waitKey()
cv2.imwrite('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/negative_image.jpg', negative_img)
```

```python
# 2. Log Transformation - enhances dark pixels using s = c * log(1 + r)
img = cv2.imread('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/images.jpg')
img_float = np.float32(img)
c = 255 / np.log(1 + np.max(img_float))  # scale output to [0, 255]
log_img = np.uint8(c * np.log(1 + img_float))
cv2.imshow('Log Transformed Image', log_img)
cv2.waitKey()
cv2.imwrite('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/log_image.jpg', log_img)
```

```python
# 3. Power Law Transformation - increases contrast using s = r^gamma (gamma=0.4)
img = cv2.imread('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/images.jpg')
gamma = 0.4
power_img = np.uint8(np.power(img / 255.0, gamma) * 255)
cv2.imshow('Power Law Image (gamma=0.4)', power_img)
cv2.waitKey()
cv2.imwrite('C:/Users/fssm0/OneDrive/Desktop/Jupyter IP/power_law_image.jpg', power_img)
```
